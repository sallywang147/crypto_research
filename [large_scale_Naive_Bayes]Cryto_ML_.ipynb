```python
encrypt_data = pd.read_csv('no_hw_binary.txt?dl=0', sep=',', header=None)
random_data = pd.read_csv('rand_no_hw_binary.txt?dl=0', sep=',',  header=None)
```

```python
#we create a list of 1's that confirms the datapoints generated by encryption
#encryption scheme is true

#we create a list a 0's that confirms the randomized datapoints are indeed randomized 
#we'll append 1's and 0's to the data enctry based on how the entry is generated 

#we do so, because it's eaisier for classifier 
enc_df = pd.DataFrame()
non_enc_df = pd.DataFrame()
encrypted = np.ones((len(encrypt_data),), dtype=int)
randomized = np.zeros((len(random_data),), dtype=int)
enc_df['Encrypted'] = encrypted.tolist()
non_enc_df['Encrypted'] = randomized.tolist()
```

```python
def name_columns(df):
  count = 0
  for index in range(513):
    count += 1
    df = df.rename(columns={ df.columns[index]: 'Vector'+str(count)})
  df.columns = [*df.columns[:-1], 'Encrypted']
  return df
```

```python
def vectorize_columns(df):
  vectorized_df = pd.DataFrame()
  vectorized_df['Vector'] = df.values.tolist()
  return vectorized_df
```

```python
from sklearn.preprocessing import StandardScaler
def standardize(df):
  std_df = df.copy(deep=True)
  scaler = StandardScaler()
  scaler.fit(std_df)
  std_df = pd.DataFrame(scaler.transform(std_df))
  return std_df
```

```python
std_enc = standardize(encrypt_data)
std_rand= standardize(random_data)
```

```python
#temp = encrypt_data.append(random_data, ignore_index=True)
#std_df = standardize(temp)
encrypted = enc_df.append(non_enc_df,ignore_index=True)
df1 = pd.merge(std_enc, enc_df, left_index=True, right_index=True)
df2 = pd.merge(std_rand, non_enc_df, left_index=True, right_index=True)
df = df1.append(df2, ignore_index=True)
df = name_columns(df)
#df = pd.merge(std_df, encrypted, left_index=True, right_index=True)
df
```

```python
from sklearn.tree import DecisionTreeClassifier
def trainer(x_train, y_train):
  model = DecisionTreeClassifier()
  model.fit(x_train, y_train)         
  return model
```

```python
def evaluate_model(df, trainer_fn, training_sizes, num_runs):
  result = np.zeros(len(training_sizes))
  predicted_label = 'Encrypted'
  for i in range(0, len(training_sizes)):
    x_train, x_test, y_train, y_test = train_test_split(
        df.drop([predicted_label], axis=1), 
        df[predicted_label], 
        test_size=0.2,
        shuffle=True,
        random_state=100)  # fixing random seed for eval at the same training size     
    for run in range(0, num_runs):  
      x_train_subset, _, y_train_subset, _ = train_test_split(
          x_train, y_train,
          train_size=training_sizes[i],
          shuffle=True,
          random_state=None)             
      model = trainer_fn(x_train_subset, y_train_subset)
      y_pred = model.predict(x_test) 
      accuracy = accuracy_score(y_pred, y_test)    
      result[i] = result[i] + accuracy
    result[i] = result[i] / num_runs            
    print('training size: %i, test accuracy: %.4f' % 
           (training_sizes[i],  result[i]))              
  return result
```

```python
training_sizes = [1000, 10000, 25000, 40000,  60000, 80000]
num_runs = 1
result = evaluate_model(df, trainer, 
                              training_sizes, num_runs)
```

```python
import plotly.express as px
print("this is the accuracy score: \n", result)
result_df = pd.DataFrame()
result_df['train_size'] = training_sizes
result_df['accuracy'] = result
px.line(result_df, x="train_size", y="accuracy", log_x=True)
```
